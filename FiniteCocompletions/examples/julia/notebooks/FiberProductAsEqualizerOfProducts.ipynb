```julia
using CapAndHomalg
```

```julia
LoadPackage( "FunctorCategories" )
```

```julia
q = FinQuiver( "q(A,B,C)[f:A->C,g:B->C]" )
```

```julia
C = PathCategory( q )
```

```julia
S = SyntacticCategoryInDoctrines( g"IsFiniteCompleteCategory", name = g"S", category = C )
```

```julia
f = S.f
```

```julia
g = S.g
```

```julia
pullback = FiberProduct( f, g )
```

```julia
program_implemented = LambdaAbstraction( pullback, ConvertJuliaToGAP( [ S, S.f, S.g ] ) )
```

```julia
Display( program_implemented )
```

```julia
svg_string = DotToSVG( DotVertexLabelledDigraph( pullback ) );
```

```julia
display( MIME("image/svg+xml"), GAPToJulia( svg_string ) )
```
