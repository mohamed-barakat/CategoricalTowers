```julia
using CapAndHomalg
```

```julia
LoadPackage( "FiniteCocompletions" )
```

```julia
q = FinQuiver( "q(a,b,c)[]" )
```

```julia
C = PathCategory( q )
```

```julia
DC = FreeDistributiveCategoryWithStrictProductAndCoproducts( C )
```

```julia
a = DC.a
```

```julia
b = DC.b
```

```julia
c = DC.c
```

```julia
axb_u_axc = Coproduct( DirectProduct( a, b ), DirectProduct( a, c ) )
```

```julia
a_x_buc = DirectProduct( a, Coproduct( b, c ) )
```

```julia
hom_axb_u_axc_a_x_buc = MorphismsOfExternalHom( axb_u_axc, a_x_buc )
```

```julia
Length( hom_axb_u_axc_a_x_buc ) == 1
```

```julia
lfactor = hom_axb_u_axc_a_x_buc[1]
```

```julia
IsOne( lfactor )
```

```julia
B = SyntacticCategoryInDoctrines( g"IsDistributiveCategory", name = g"B", category = C )
```

```julia
Display( B )
```

```julia
F = EmbeddingOfUnderlyingCategory( B )
```

```julia
DF = ExtendFunctorToFreeDistributiveCategoryWithStrictProductAndCoproducts( F )
```

```julia
blfactor = DF( lfactor )
```

```julia
arguments = [ B, B.a, B.b, B.c ]
```

```julia
Display( blfactor )
```

```julia
program_generated = LambdaAbstraction( blfactor, ConvertJuliaToGAP(arguments) )
```

```julia
display( MIME("image/svg+xml"), GAPToJulia( DotToSVG( DotVertexLabelledDigraph( blfactor ) ) ) )
```

```julia
Display( program_generated )
```

```julia
program_implemented = LambdaAbstraction( LeftCartesianDistributivityFactoring( B.a, [ B.b, B.c ] ), ConvertJuliaToGAP(arguments) )
```

```julia
Display( program_implemented )
```
